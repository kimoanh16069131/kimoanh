```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
sns.relplot(x="total_bill",y="tip", data=tips)
```

```
sns.relplot(x="total_bill",y="tip", data=tips, hue="smoker")
```

```
sns.relplot(x="total_bill",y="tip", data=tips, hue="smoker",style="smoker")
```

```
sns.relplot(x="total_bill",y="tip", data=tips, hue="smoker",style="time")
```

```
sns.relplot(x="total_bill",y="tip", data=tips, style="size")
```

```
sns.relplot(x="total_bill",y="tip", data=tips, size="size",sizes=(15,200))
```

```
sns.relplot(x="total_bill",y="tip", data=tips, hue="smoker", col="time")
```

```
sns.regplot(x="total_bill",y="tip", data=tips)
```

```
sns.lmplot(x="total_bill",y="tip", data=tips, hue="smoker", col="time")
```

```
sns.lmplot(x="total_bill",y="tip", data=tips, hue="smoker", col="time",row="sex")
```
